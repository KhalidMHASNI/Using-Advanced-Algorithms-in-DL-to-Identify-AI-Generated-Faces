```python
# --- Importing necessary libraries and modules ---
from preprocessing.filters import apply_all_filters  # Import the function that applies all predefined filters to an image
from preprocessing.patch_generator import smash_n_reconstruct  # Import the function to generate rich and poor texture patches from an image
import matplotlib.pyplot as plt  
import numpy as np  
```

### Processing AI generated Image

```python
# --- Extracting rich and poor texture patches from an AI-generated image ---
rt1, pt1 = smash_n_reconstruct('../dataset/fake/00bc43b8-4151-4a6f-9370-002f2f3e918d.jpg')  # Extract rich texture (rt1) and poor texture (pt1) from an AI-generated image

# --- Visualizing the rich and poor textures of the AI-generated image ---
plt.subplot(1,2,1)  # Create a subplot with 1 row and 2 columns, plot on the first subplot
plt.imshow(rt1)  # Display the rich texture image
plt.title('rich texture - AI generated image')  # Title for the rich texture subplot

plt.subplot(1,2,2)  # Plot on the second subplot
plt.imshow(pt1)  # Display the poor texture image
plt.title('poor texture - AI generated')  # Title for the poor texture subplot

plt.show()  # Show the combined plot for both textures
```

### Applying high pass filters

```python
# --- Apply high pass filters to the rich and poor textures of the AI-generated image ---
rt1 = apply_all_filters(rt1)  # Apply the filter to the rich texture
pt1 = apply_all_filters(pt1)  # Apply the filter to the poor texture
```

```python
# --- Visualizing the filtered rich and poor textures of the AI-generated image ---
plt.subplot(1,2,1)  # Create a subplot with 1 row and 2 columns, plot on the first subplot
plt.imshow(rt1, cmap='gray')  # Display the filtered rich texture in grayscale
plt.title('rich texture - AI generated image')  # Title for the rich texture subplot

plt.subplot(1,2,2)  # Plot on the second subplot
plt.imshow(pt1, cmap='gray')  # Display the filtered poor texture in grayscale
plt.title('poor texture - AI generated')  # Title for the poor texture subplot

plt.show()  # Show the combined plot for both texture
```
